```python
import matplotlib.pyplot as plt
%matplotlib widget
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import make_blobs
from IPython.display import display, Markdown, Latex
import logging
logging.getLogger("tensorflow").setLevel(logging.ERROR)
tf.autograph.set_verbosity(0)
from matplotlib.widgets import Slider
```

```python
def softmax(z):
    ez = np.exp(z)
    sm = ez/np.sum(ez)
    return sm
plt.close("all")
```

```python
centers = [[-5,2],[-2, -2], [1, 2], [5, -2]]
X_train, y_train = make_blobs(n_samples=2000, centers=centers, cluster_std=1.0,random_state=30)
```

```python
#non-preferred
model = Sequential( 
    [
    Dense(25, activation='relu'),
    Dense(15, activation='relu'),
    Dense(4, activation='softmax')
    ]
)
model.compile(
    loss = tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer =  tf.keras.optimizers.Adam(0.001),
)
model.fit(
        X_train,y_train,epochs=10
)
```

```python
ans = model.predict(X_train)
print(ans[:2])
print("largest",np.max(ans),"smallest",np.min(ans))
```

```python
#preferred implimentation
model = Sequential(
    [
        Dense(25, activation='relu'),
        Dense(15, activation='relu'),
        Dense(4, activation='linear'),
        ]
)
model.compile(
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer = tf.keras.optimizers.Adam(0.001),
)
model.fit(
    X_train, y_train, epochs=10
)
```

```python
ans = model.predict(X_train)
print(f"two output vectors are {ans[:2]}")
print("Largest value",np.max(ans), "Smallest:", np.min(ans))
```

```python
#to get probability form
prob = tf.nn.softmax(ans).numpy()
print(f"two output vectors are {prob[:2]}")
print("Largest value",np.max(prob), "Smallest:", np.min(prob))
```

```python
#to find most likely
for i in range(5):
    print(f"{ans[i]}, category:{np.argmax(ans[i])}")
```
